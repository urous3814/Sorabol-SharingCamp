```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/colab/202108_202108_연령별인구현황_월간 (1).csv', encoding='cp949', thousands = ',')
df
```

```python
df = df[df['2021년08월_계_총인구수'] != 0]
df
```

```python
df[df['행정구역'].str.contains('노원구')]
```

```python
tmp = df[df['행정구역'].str.contains('노원구')]
tmp2 = tmp.loc[:,'2021년08월_계_0세':'2021년08월_계_100세 이상'] 
tmp3 = tmp2.to_numpy()
town_np = tmp3[0]
town_np
```

```python
import matplotlib.pyplot as plt 
plt.title('노원구의 인구 구조', fontsize=25)
plt.plot(town_np) 
plt.show() 
```

```python
M_np = df[df['행정구역'].str.contains('노원구')].loc[:,'2021년08월_남_0세':'2021년08월_남_100세 이상'].to_numpy()[0]
F_np = df[df['행정구역'].str.contains('노원구')].loc[:,'2021년08월_여_0세':'2021년08월_여_100세 이상'].to_numpy()[0]
plt.title('노원구의 남녀 인구 구조', fontsize=25)
plt.plot(M_np, label='남자')
plt.plot(F_np, label='여자')
plt.legend()
plt.show()
```

```python
data = M_np - F_np
plt.figure(figsize=(10, 5))
plt.title('노원구의 남자 여자의 인구수 차이 (남자-여자)', fontsize=25)
plt.bar(range(101), data)
plt.show()
```

```python
M2_np = df[df['행정구역'].str.contains('1100000000')].loc[:, '2021년08월_남_0세':'2021년08월_남_100세 이상'].to_numpy()[0]
F2_np = df[df['행정구역'].str.contains('1100000000')].loc[:, '2021년08월_여_0세':'2021년08월_여_100세 이상'].to_numpy()[0]
data2 = M2_np - F2_np
plt.figure(figsize=(10, 6))
plt.title('서울특별시의 남자 여자의 인구수 차이 (남자-여자)', fontsize=25)
plt.bar(range(101), data2)
plt.plot(range(101), data, 'r')
plt.show()
```

```python
M3_np = df[df['행정구역'].str.contains('성북구')].loc[:, '2021년08월_남_0세':'2021년08월_남_100세 이상'].to_numpy()[0]
F3_np = df[df['행정구역'].str.contains('성북구')].loc[:, '2021년08월_여_0세':'2021년08월_여_100세 이상'].to_numpy()[0]
data3 = M3_np - F3_np
plt.figure(figsize=(10, 6))
plt.title('성북구의 남자 여자의 인구수 차이 (남자-여자)', fontsize=25)
plt.bar(range(101), data3)
plt.plot(range(101), data, 'r')
plt.show()
```
