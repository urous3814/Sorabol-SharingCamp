```python
df2 = df.loc[:, '행정구역':'2021년08월_계_100세 이상']
df2.head()
```

```python
for col in range(3, 104) :
   df2.iloc[:,col] = df2.iloc[:,col] / df2.iloc[:,1]
df2.head()
```

```python
town_df = df2[df2['행정구역'].str.contains('노원구')]
town_df
```

```python
list_mse = []
for i in range(len(df2)):
  list_mse.append(((df2.iloc[i,3:] - town_df.iloc[0,3:])**2).mean())
df2['MSE'] = list_mse
df2.head()
```

```python
df3 = df2.sort_values('MSE')
df3.head()
```

```python
plt.figure(figsize = (10, 6))
plt.title('노원구와 가장 유사한 인구구조를 가진 지역과 가장 먼 인구구조를 가진 지역만을 찾아왔으나 이제는 더이상 존재하지 않으니 이를 어찌할꼬...', fontsize = 25)
plt.plot(df3.iloc[0,3:104].to_numpy(), label = df3.iloc[0,0])
plt.plot(df3.iloc[1,3:104].to_numpy(), label = df3.iloc[1,0])
plt.plot(df3.iloc[-1,3:104].to_numpy(), label = df3.iloc[-1,0])
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.title('노원구와 가장 유사한 인구구조를 가진 지역과 가장 먼 인구구조를 가진 지역', fontsize = 25)
plt.plot(df3.iloc[0,3:104].to_numpy(), label = df3.iloc[0,0].split('(')[0])
plt.plot(df3.iloc[1,3:104].to_numpy(), label = df3.iloc[1,0].split('(')[0])
plt.plot(df3.iloc[-1,3:104].to_numpy(), label = df3.iloc[-1,0].split('(')[0])
plt.legend()
plt.show()
```
